```python
import pygmsh
import meshio
import os
import sys
sys.path.insert(0, '/root/shared/')
import visualisation_helpers, tetridiv
```

```python
show_args = {'jupyter_backend': 'panel', 'return_viewer': True}
```

# 2d Surface Tests

```python
centre = [0.0, 0.0, 0.0]
radius = 1.0
mesh_size = 0.1
with pygmsh.geo.Geometry() as geom:
    geom.add_circle(centre, radius, mesh_size, num_sections = 4)
    pygmsh_mesh = geom.generate_mesh(dim=2)
    pygmsh.write('circle.msh')
mesh = meshio.read("circle.msh")
os.remove("circle.msh")
```

```python
divmesh = tetridiv.tri2quad(mesh)
```

```python
p = visualisation_helpers.plot_mesh(mesh)
p.show(**show_args)
```

```python
p = visualisation_helpers.plot_mesh(divmesh)
p.show(**show_args)
```

# 3d Surface Test

```python
centre = [0.0, 0.0, 0.0]
radii = [1.0, 1.5, 2.0]
mesh_size = 0.1
with pygmsh.geo.Geometry() as geom:
    geom.add_ellipsoid(centre, radii, mesh_size, with_volume=False)
    pygmsh_mesh = geom.generate_mesh(dim=2)
    pygmsh.write('ellipsoid.msh')
mesh = meshio.read("ellipsoid.msh")
os.remove("ellipsoid.msh")
```

```python
divmesh = tetridiv.tri2quad(mesh)
```

```python
p = visualisation_helpers.plot_mesh(mesh, slice_normal='x')
p.show(**show_args)
```

```python
p = visualisation_helpers.plot_mesh(divmesh, slice_normal='x')
p.show(**show_args)
```
